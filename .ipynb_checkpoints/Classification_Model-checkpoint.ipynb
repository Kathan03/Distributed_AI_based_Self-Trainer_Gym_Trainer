### Data Preprocessing

```python
dataset_df = pd.DataFrame(columns=["Sequences", "vid_id"])
for id in angle_df['vid_id'].unique():
    X_sequence = np.array(angle_df[angle_df['vid_id'] == id].iloc[:, 2:9])
    new_data = [[list(X_sequence.tolist()), id]]
    new_df = pd.DataFrame(new_data, columns=["Sequences", "vid_id"])
    dataset_df = pd.concat([dataset_df, new_df], ignore_index=True)
dataset_df
```

```python
label_df = label_df.sort_values(by='vid_id')
dataset_df = dataset_df.sort_values(by='vid_id')
# dataset_df.drop('class')
dataset_df['category'] = label_df['class']
```

```python
dataset_df
```

```python
def category_to_index(category):
    possible_labels=dataset_df['category'].unique()
    label_dict = {}
    for index, label in enumerate(possible_labels):
        label_dict[label] = index
    #print(label_dict)
    return label_dict[category]

dataset_df['class'] = dataset_df['category'].apply(category_to_index)
dataset_df
```

### Training

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(dataset_df['Sequences'], dataset_df['class'], test_size = 0.2)
```

```python
y_train = np.asarray(y_train)
```

```python
padded_train = pad_sequences(x_train, maxlen = 350, padding = 'post', truncating = 'post')
padded_test = pad_sequences(x_test, maxlen = 350, padding = 'post', truncating = 'post')

# import torch
# padded_train = torch.nn.utils.rnn.pad_sequence(x_train, batch_first=True, padding_value=0.0)
Xtrainshape = padded_train.shape
Xtrainshape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding, Input, LSTM, Conv1D, MaxPool1D, Bidirectional
from tensorflow.keras.models import Model


#Sequential Model

model = Sequential()
# model.build(input_shape = padded_train.shape) 
#Bi-Directional RNN/LSTM
model.add(Bidirectional(LSTM(256, return_sequences=True, dropout=0.3, recurrent_dropout=0.2),input_shape=(Xtrainshape[1],Xtrainshape[2])))
#model.add(Bidirectional(LSTM(200, return_sequences=True, dropout=0.25, recurrent_dropout=0.15)))
model.add(Bidirectional(LSTM(64, return_sequences=True, dropout=0.2, recurrent_dropout=0.1)))
#model.add(Bidirectional(LSTM(32)))
model.add(Bidirectional(LSTM(16)))

#Dense Layer
model.add(Dense(1028, activation = 'relu'))
model.add(Dense(720, activation = 'relu'))
model.add(Dense(512, activation = 'relu'))
model.add(Dense(256, activation = 'relu'))
#model.add(Dense(256, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(5, activation = 'softmax'))
model.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics=['acc'])
model.summary()
```

```python
# Initializing callbacks
earlystopping = EarlyStopping(monitor = 'val_loss', mode = 'min', verbose = 1, patience = 20)
checkpointer = ModelCheckpoint("model_weights.h5", monitor='val_accuracy', save_weights_only=True, mode='max', verbose = 1)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor = 0.1, patience = 2, min_lr = 0.00001, model='auto')

callbacks = [checkpointer, reduce_lr, earlystopping]
```

```python
history = model.fit(padded_train, y_train, batch_size = 10, validation_split = 0.3, epochs = 2, callbacks=callbacks)
```

```python
accuracy = history.history['acc']
val_accuracy  = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(accuracy))
plt.plot(epochs, loss, 'C1', label = 'Training_loss')
plt.plot(epochs, val_loss, 'C2', label = 'Validation_loss')
plt.legend()
```

```python
plt.plot(epochs, accuracy, 'C2', label = 'Training_accuracy')
plt.plot(epochs, val_accuracy, 'C3', label = 'Validation_accuracy')
plt.legend()
```

```python
model.save('my_model.keras')
model.load_weights('model_weights.h5')
```

```python
pred1 = model.predict(padded_test)
```

```python
#Segmenting prediction of model into classes 
prediction = []
for i in range(len(pred1)):
    label = np.argmax(pred1[i])
    prediction.append(label)
```

```python
#Estimating the Accuracy based on test input

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(list(y_test), prediction)

print("Model Accuracy : ", accuracy)
```

### Landmarlks to angles

```python
df_cor = pd.read_csv('./landmarks.csv')
df_cor
```

```python
right_elbow = [df_cor['x_right_elbow'][1], df_cor['y_right_elbow'][1], df_cor['z_right_elbow'][1]] 
right_shoulder = [df_cor['x_right_shoulder'][1], df_cor['y_right_shoulder'][1], df_cor['z_right_shoulder'][1]]
right_wrist = [df_cor['x_right_wrist'][1], df_cor['y_right_wrist'][1], df_cor['z_right_wrist'][1]]
```

```python
def calc_angle(a,b,c):
    a = np.array(a)
    b = np.array(b)
    c = np.array(c)

    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)
    angle = np.degrees(angle)

    if angle > 180.0:
        angle = 360-angle
    return angle
```

```python
angle = calc_angle(right_shoulder, right_elbow, right_wrist)
print(angle)
```
